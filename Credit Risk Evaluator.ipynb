```python
import numpy as np
import pandas as pd
from pathlib import Path
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))

train_df['debt_settlement_flag'].nunique()
test_df['debt_settlement_flag'].nunique()
```

```python
# Convert categorical data to numeric and separate target feature for training data
X = train_df.drop('loan_status', axis=1)


X_train_dummies = pd.get_dummies(X)


y_label_train = LabelEncoder().fit_transform(train_df['loan_status'])

X_train_dummies
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X = test_df.drop('loan_status', axis=1)

X_test_dummies = pd.get_dummies(X)


y_label_test = LabelEncoder().fit_transform(test_df['loan_status'])

print(X_test_dummies.columns)
print()
print(X_train_dummies.columns)
```

```python
# add missing dummy variables to testing set
X_train_dummies['debt_settlement_flag_Y']

X_train_dummies
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier

classifier.fit(X_train_dummies, y_label_train)


print(f"Training Data Score: {classifier.score(X_train_dummies, y_label_train)}")
print(f"Testing Data Score: {classifier.score(X_test_dummies, y_label_test)}")
```
